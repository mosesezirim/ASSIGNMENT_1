```python
import pandas as pd
file = r'C:\Users\moses\Desktop\data\shifts.xlsx'
fm = pd.read_excel(file)
fm.head()
```

```python
#.3-Using the shifts.xlsx data, find the total Units sold for each of the regions
fm.groupby('Region')['Units Sold'].sum()
```

```python
ms = fm.groupby('Sales Rep')['Units Sold'].sum()
ms.head()
```

```python
# 4-Find the top three sales reps that had the highest units sold in total, also listing the total units sold by each of them.
ms.nlargest(3)
```

```python
# 5-Using visualization, find the product with the highest unit sold and state the units sold.
import matplotlib.pyplot as plt
plt.bar(fm['Product'], fm['Units Sold'])
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.title('PRODUCT CHART')
plt.show()
```

```python
#Similarly, using visualization, find the region with the highest units sold
plt.bar(fm['Region'], fm['Units Sold'])
plt.xlabel('Region')
plt.ylabel('Units Sold')
plt.title('REGION CHART')
plt.show()
```

```python
#Create a visualization to examine the shift with the highest units sold
plt.bar(fm['Shift'], fm['Units Sold'])
plt.xlabel('Shift')
plt.ylabel('Units Sold')
plt.title('SHIFT CHART')
plt.show()
```
